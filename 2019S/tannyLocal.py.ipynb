```python
import cv2
import numpy as np
import os
dirname = 'export'




def Number(img,Area,Height):
    #img = cv2.imread("test6.jpg")
    img = cv2.resize(img,(600,400))

    #Greyscale
    grey = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

    #Blur
    blur = cv2.GaussianBlur(grey,(5,5),2,3)
    ##Gabor kernel for tilted 

    #Edge 
    edge = cv2.Canny(blur, 70,70)

    #binary
    thresh = cv2.adaptiveThreshold(blur,127,1,1,11,2)

    # find contour
    img_cnt ,contours, hierarchy = cv2.findContours(thresh, cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_NONE)

    #draw rectangle

    boxes=[]
    num = 1
    for cnt in contours:
        area = cv2.contourArea(cnt)
        
        if area > Area:
            [x,y,w,h]= cv2.boundingRect(cnt)
            if h > Height:
                boxes.append([x,y,w,h])
                
                #function to resize
                i = resize2SquareKeepingAspectRation(grey[y-15:y+h+15,x-15:x+w+15],28, cv2.INTER_AREA)
                
                # saving all files
                cv2.imwrite(os.path.join(dirname,"image_{}.jpg".format(num) ), i)
                
                #cv2.imwrite("image_{}.jpg".format(num),i)
                num+=1
                
                #draw bounding boxes
                cv2.rectangle(img,(x-15,y-15),(x+w+15,y+h+15),[0,255,0],2)
                cv2.imshow("img",img)
                cv2.waitKey(0)
                cv2.destroyAllWindows()
    return (img)




def resize2SquareKeepingAspectRation(img, size, interpolation):
    
    h, w = img.shape[:2]
    c = None if len(img.shape) < 3 else img.shape[2]
    if h == w: return cv2.resize(img, (size, size), interpolation)
    if h > w: dif = h
    else:     dif = w
        
    x_pos = int((dif - w)/2.)
    y_pos = int((dif - h)/2.)
    
    if c is None:
        mask = np.ones((dif, dif), dtype=img.dtype)
        mask[y_pos:y_pos+h, x_pos:x_pos+w] = img[:h, :w]
    else:
        mask = np.ones((dif, dif, c), dtype=img.dtype)
        mask[y_pos:y_pos+h, x_pos:x_pos+w, :] = img[:h, :w, :]
        
    return cv2.resize(mask, (size, size), interpolation)




# test stage-----------------------------------------------------------------------

im = cv2.imread("test4.jpg")
im_test = Number(im,30,40)
```
